```python
import numpy as np
import matplotlib.pyplot as plt

@np.vectorize
def entropy(*vec):
    # takes in a vector that sums to 1
    total = 0
    for item in vec:
        total += np.log2(item)*item
    return -total

dx, dy = 0.01, 0.01

# generate 2 2d grids for the x & y bounds
y, x = np.mgrid[slice(0, 1, dy),
                slice(0, 1, dx)]
z = 1-y-x
c = entropy(x,y,z)
print(c)
```

```python
plt.pcolor(x, y, c, cmap='RdBu')
plt.title('pcolor')
plt.colorbar()
```
